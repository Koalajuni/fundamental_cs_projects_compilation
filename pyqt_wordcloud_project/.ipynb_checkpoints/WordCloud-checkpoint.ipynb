### 순서

1. 데이터 수집 
2. 데이터를 사용할 수 있도록 하나의 변수로 처리 예) 여러개의 Dataframe을 1개의 변수로 
3. 불필요한 데이터를 제거 및. 필요한 데이터 추가 
4. pandas 데이터를 하나의 문자열로 변환

```python
from collections import Counter
import numpy as np
import pandas as pd 
import seaborn as sns
import re
import matplotlib.pyplot as plt
import os
from wordcloud import WordCloud

```

```python
# 특정 경로에 있는 CSV파일만 불러오기
path = "./search_file/"
```

```python
file_list = os.listdir(path)
print(file_list)
```

```python
fileList1,fileList2 = [], []
aiList = ['인공지능.csv', 'AI.csv','챗gpt.csv']
btcList = ['코인.csv','비트코인.csv', '비트코인 가격.csv']
for file in file_list:
    if file in aiList:
        fileList1.append(file)
    if file in btcList:
        fileList2.append(file)
```

```python
dfCsv1,dfCsv2 = pd.DataFrame(),pd.DataFrame()

for i in fileList1:
    data = pd.read_csv(path + i, encoding ='utf-8')
    dfCsv1 = pd.concat([dfCsv1,data])

for i in fileList2:
    data = pd.read_csv(path + i, encoding ='utf-8')
    dfCsv2 = pd.concat([dfCsv2,data])

dfCsv1 = dfCsv1.reset_index(drop = True)
dfCsv2 = dfCsv2.reset_index(drop = True)
```

#### 불필요한 데이터 제거

```python
dfCsv1 = dfCsv1.drop(['링크'], axis = 'columns')
dfCsv2 = dfCsv2.drop(['링크'], axis = 'columns')
```

## 전처리
#### 1. 월드클라우드를 위해 Rows를 하나의 문장으로 변환 -> pandas to one word
#### 2. 특수문자를 제거하고, 오직 한문,영문 글자로만 변경 -> Regex

```python
print(dfCsv1.head())
```

```python
listCsv1 = dfCsv1['제목'].astype(str).tolist()
listCsv2 = dfCsv2['제목'].astype(str).tolist()

oneWord1 = (' ').join(re.sub('[\W_]+', ' ', word) for word in listCsv1)
oneWord2 = (' ').join(re.sub('[\W_]+', ' ', word) for word in listCsv2)
```

```python
def clean_text(inputString):
    text_rmv = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’\'·]', ' ', inputString)
    return text_rmv
```

#### 워드클라우드에서 키워드와 비슷한 단어들을 정규 표현식으로 미리 제거

```python
def clean_key(inputString):
    text_rmv = re.sub('인공지능?', ' ', inputString)
    text_rmv = re.sub('[AI]', ' ', inputString)
    text_rmv = re.sub('챗GPT?', ' ', inputString)
    text_rmv = re.sub('Chat?', ' ', inputString)
    text_rmv = re.sub('GPT?', ' ', inputString)
    return text_rmv

oneWord1 = clean_key(oneWord1)
```

### wordcloud 실행하기 

- 월드클라우드 실행
- 각 단어별로 카운트 Dict 생성 
- plot으로 보여주기

```python
stopList1 = ["인공지능","챗GPT","챗 GPT","챗", "어떻게", "질문","AI","인공지능이", "AI가","관련","이","의","학과","과"] 
stopList2 = ["비트코인","코인","Bitcoin","암호화폐","가격","왜","의","학과","과","질문","거래소","업비트","관련"]
wc1 = WordCloud(
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc",
stopwords = stopList1, 
background_color = "white",
width = 700, height = 700,
random_state = 40)

wc2 = WordCloud(
font_path = "/System/Library/Fonts/AppleSDGothicNeo.ttc",
stopwords = stopList2, 
background_color = "white",
width = 700, height = 700,
random_state = 40)

cloud1 = wc1.generate(oneWord1)
cloud2 = wc2.generate(oneWord2)
```

```python
#matplotlib으로 보여주기 

plt.figure(figure = (10,8))
plt.axis('off')
plt.imshow(cloud2)
plt.show()
```
